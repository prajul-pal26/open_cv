EVENTS - draw a drawing after cicking on the window 

```python
img = np.zeros((512,512,3))
def DrawShape(event, x, y, flags, param):
    # event ka type(3- move, click, ideal), x, y, flags, param
    """
    0 = movement of the mouse
    1 = click
    4 = ideal
    """
    if event == 1:
        cv2.circle(img,center=(x,y),radius=40, color=(0,255,255),thickness=-1)
            
cv2.namedWindow('Event')   # jis window me kaam hoga 
cv2.setMouseCallback('Event', DrawShape)    
while True:
    cv2.imshow('Event', img)
    if cv2.waitKey(1) & 0xFF ==ord('x'):
        
        break
plt.imshow(img)
cv2.destroyAllWindows()
```

```python
img= cv2.imread('image.png')
original = img
img = cv2.resize(img, (800, 600))

flag = False
ix =-1
iy =-1
finished = False
def CropImage(event, x, y, flags, param):
    
    """
    0 = movement of the mouse
    1 = click
    4 = ideal
    """
    global ix, iy, flag,finished
    # if event==0:
    #     if flag==True:
    #         cv2.boxPoints(img,points=)
            # cv2.rectangle(img, pt1=(ix,iy),pt2=(x,y), color=(0,0,255), thickness=1)
    if event==1:
        flag = True
        ix = x
        iy= y
    if event==4:
        if flag==True:

            cv2.rectangle(img, pt1=(ix,iy),pt2=(x,y), color=(0,0,255), thickness=1)
            crop = img[iy:y,ix:x]            
            
            crop = cv2.cvtColor(crop, cv2.COLOR_BGR2RGB)
            
            

            fig, axs = plt.subplots(1, 3, figsize=(15, 5))

            axs[0].imshow(original)
            axs[0].set_title("ORIGINAL")
            axs[0].axis('off')

            axs[1].imshow(img)
            axs[1].set_title("Processing")
            axs[1].axis('off')
            
            axs[2].imshow(crop)
            axs[2].set_title("Final Image")
            axs[2].axis('off')
            plt.tight_layout()
            plt.show()
            finished = True
            
            
        flag=False 
        
cv2.namedWindow('Photo')
cv2.setMouseCallback('Photo', CropImage)


while True:
    cv2.imshow('Photo', img)
    if finished:
        break
    if cv2.waitKey(1) & 0xFF ==ord('x'):
        break
    
cv2.destroyAllWindows()
```
